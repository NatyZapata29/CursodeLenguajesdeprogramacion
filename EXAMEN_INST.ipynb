**CALCULAR EL VALOR NETO**

```python
# Función para calcular el VPN de un proyecto
def calcular_vpn(flujos, tasa_descuento):
    vpn = 0
    for t in range(len(flujos)):
        vpn += flujos[t] / ((1 + tasa_descuento) ** t)
    return vpn

# Definir la tasa de descuento (12%)
tasa = 0.12

# Asegurarse de que los nombres de las columnas coincidan con los esperados
columnas_flujo = ['Inversion_Inicial'] + [f'Año{i}' for i in range(1, 11)]

# Calcular el VPN para cada proyecto
df['VPN'] = df.apply(
    lambda fila: calcular_vpn([fila[col] for col in columnas_flujo], tasa),
    axis=1
)

# Ver resultados
df[['Capitalizador', 'VPN']].head()
```

```python
import numpy_financial as npf  # Importar librería financiera

# Calcular la TIR para cada proyecto
df['TIR'] = df.apply(
    lambda fila: npf.irr(
        [fila['Inversion_Inicial']] + [fila[f'Año{i}'] for i in range(1, 11)]
    ),
    axis=1
)

# Mostrar resultados
from IPython.display import display
display(df[['Capitalizador', 'VPN', 'TIR']])
```

**GRAFICAS PARA VISUALIZAR MEJOR VPN

```python

import matplotlib.pyplot as plt

# Filtrar los 10 proyectos con mayor VPN
top_10_proyectos_vpn = df.nlargest(10, 'VPN')

# Gráfico de barras: Proyectos vs VPN para los 10 proyectos con mayor VPN
plt.figure(figsize=(12, 6))
bars = plt.bar(top_10_proyectos_vpn['Capitalizador'], top_10_proyectos_vpn['VPN'], color='green')

# Titulos de todos los ejes
plt.xlabel('Capitalizador')
plt.ylabel('Valor Presente Neto (VPN)')
plt.title('Top 10 Proyectos con Mayor VPN')

plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Eliminar etiquetas de las barras apiladas
plt.tight_layout()
plt.show()
```

**GRAFICAS PARA TIR**

```python
# Filtrar proyectos con TIR mayor a la tasa de descuento
tasa_descuento = 0.12
proyectos_rentables = df[df['TIR'] > tasa_descuento]

# Gráfico de barras: Inversionista vs VPN
plt.figure(figsize=(10, 6))
plt.bar(proyectos_rentables['Capitalizador'], proyectos_rentables['VPN'], color='green')

plt.xlabel('Capitalizador')
plt.ylabel('Valor Presente Neto (VPN)')
plt.title('Proyectos con TIR mayor al 12%')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```

**SUSTENTACION**

Se llevó a cabo un ejercicio utilizando un archivo CSV que contenía información de aproximadamente mil inversionistas, cada uno con una inversión inicial y proyecciones de ingresos a lo largo de diez años.

En una primera etapa, se realizó una verificación de la integridad del archivo, asegurándose de que no existieran valores nulos o celdas vacías. Posteriormente, se aplicaron métodos estadísticos básicos para obtener una visión general de la estructura del conjunto de datos.

Luego, se procedió a efectuar análisis financieros fundamentales, comenzando con el cálculo del Valor Presente Neto (VPN), una métrica que permite evaluar si una inversión es rentable. Este indicador determina si los ingresos estimados en el futuro superan el capital invertido inicialmente, teniendo en cuenta una tasa de descuento fija del 12%, seleccionada por estar por encima del promedio del IPC de la última década. Este valor se utilizó para traer los flujos de efectivo futuros al presente, facilitando así su evaluación.

Seguidamente, se calculó la Tasa Interna de Retorno (TIR), la cual representa el rendimiento porcentual de la inversión. Cuando este valor supera la tasa de descuento, se interpreta que la inversión es financieramente recomendable.
